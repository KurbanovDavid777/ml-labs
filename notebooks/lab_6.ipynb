**Пропуски**

1) там где пропусков больше 90% - удалим столбец
2) пропусков > 5% и ≤ 90% либо заполняем медианой (если число) либо unknow (если строка)
3) Пропусков ≤ 5% удаляем строки

```python
missing_percent = df.isna().mean()*100

#колонки где пропусков больше 90%
columns_to_drop = missing_percent[missing_percent > 90].index
df = df.drop(columns=columns_to_drop)
print(f"Удалены колонки с пропусками большее 90%: {list(columns_to_drop)}")

#Пропуски с 5-90% заполняем
for col in missing_percent[(missing_percent > 5) & (missing_percent < 90)].index:
 #числа
    if df[col].dtype in [np.float64, np.int64]:
      df[col] = df[col].fillna(df[col].median())

    elif col == 'box_office':
      df[col] = df[col].fillna(0)

    #категориальные
    else:
      df[col] = df[col].fillna("Unknown")

#Колоник где пропусков меньше 5%
low_missing_percent = missing_percent[missing_percent < 5].index
df = df.dropna(subset=low_missing_percent)
print(f"Удалены строки, где были пропуски в колонках с ≤5% пропусков: {list(low_missing_percent)}")
```

# Построение baseline-модели нейронной сети

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

Источник, которым вдохновился: https://stackoverflow.com/questions/72874433/neural-network-for-regression-using-pytorch

```python
X_train=pd.read_csv('/content/X_train.csv')
X_test=pd.read_csv('/content/X_test.csv')

y_train=pd.read_csv('/content/y_train.csv')
y_test=pd.read_csv('/content/y_test.csv')
```

```python
type(X_train), type(y_train)
```

```python
y_train.columns
```

```python
##### Creating Tensors
X_train_tensor=torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor=torch.tensor(X_test.values, dtype=torch.float32)

y_train_tensor=torch.tensor(y_train.values, dtype=torch.float32).view(-1,1)
y_test_tensor=torch.tensor(y_test.values, dtype=torch.float32).view(-1,1)
```

```python
class ANN_Model(nn.Module):
  def __init__(self, input_features=4, hidden1=20, hidden2=20, out_features=1):
    super().__init__()
    self.f_connected1=nn.Linear(input_features, hidden1)
    self.f_connected2=nn.Linear(hidden1, hidden2)
    self.out=nn.Linear(hidden2, out_features)

  def forward(self, x):
    x=F.relu(self.f_connected1(x))
    x=F.relu(self.f_connected2(x))
    x=self.out(x)
    return x

model = ANN_Model(input_features=X_train_tensor.shape[1])

loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)

epochs = 1000
final_losses = []

for i in range(epochs):
  y_pred = model.forward(X_train_tensor)
  loss = loss_function(y_pred, y_train_tensor)
  final_losses.append(loss.item())

  if i%10==1:
    print("Epoch number: {} and the loss: {}".format(i, loss.item()))

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F

#График изменения потерь
plt.figure(figsize=(10,5))
plt.plot(final_losses)
plt.xlabel("Эпохи")
plt.ylabel("MSE Loss")
plt.title("Изменение ошибки на тренировке")
plt.show()

# 2️MSE на тестовой выборке
model.eval()  # переводим модель в режим оценки
with torch.no_grad():  # отключаем градиенты
    y_test_pred = model(X_test_tensor)
    mse_test = F.mse_loss(y_test_pred, y_test_tensor)
    rmse_test = torch.sqrt(mse_test)
    print(f"RMSE на тестовой выборке: {rmse_test.item():.4f}")
```

```python
from pytorch_tabular import TabularModel, TabularModelTuner
from pytorch_tabular.models import (
    CategoryEmbeddingModelConfig,
)
from pytorch_tabular.config import DataConfig, OptimizerConfig, TrainerConfig, ExperimentConfig
from pytorch_tabular.models.common.heads import LinearHeadConfig
```

```python
df_scaled.info()
```

```python

numerical_cols = [
    "type", "film_studio", "production_country", "director",
    "producer", "age_restriction", "genres",
    "box_office", "box_office_log", "box_office_clipped"
]
```

```python
data_config = DataConfig(
    target=[
        "ratings"
    ],  # target should always be a list. Multi-targets are only supported for regression. Multi-Task Classification is not implemented
    continuous_cols=numerical_cols,
    categorical_cols=[],
)
```

```python
trainer_config = TrainerConfig(
    batch_size=32,
    max_epochs=50,
    early_stopping="valid_loss",  # Monitor valid_loss for early stopping
    early_stopping_mode="min",  # Set the mode as min because for val_loss, lower is better
    early_stopping_patience=5,  # No. of epochs of degradation training will wait before terminating
    checkpoints="valid_loss",  # Save best checkpoint monitoring val_loss
    load_best=True,  # After training, load the best checkpoint
    progress_bar="none",  # Turning off Progress bar
    trainer_kwargs=dict(enable_model_summary=False),  # Turning off model summary
)
optimizer_config = OptimizerConfig()
```

```python
head_config = LinearHeadConfig(
    layers="", dropout=0.1, initialization="kaiming"  # No additional layer in head, just a mapping layer to output_dim
).__dict__  # Convert to dict to pass to the model config (OmegaConf doesn't accept objects)
```

```python
model_config = CategoryEmbeddingModelConfig(
    task="regression",
    layers="20-20",  # Number of nodes in each layer
    activation="ReLU",  # Activation between each layers
    learning_rate=1e-3,
    head="LinearHead",  # Linear Head
    head_config=head_config,  # Linear Head Config
)
```

```python
search_space = {
    'model_config__activation': ["ReLU", "LeakyReLU", "ELU"],           # функция активации
    'model_config__dropout': [0.0, 0.1, 0.3, 0.5],         # Dropout
    'model_config__use_batch_norm': [True, False],                      # Batch Normalization                         # Размер батча
}
```

```python
import warnings
```

```python
import torch
import collections
from omegaconf import DictConfig
from omegaconf.listconfig import ListConfig
from omegaconf.nodes import AnyNode
from omegaconf.base import Metadata

# Добавляем все безопасные типы для unpickling
torch.serialization.add_safe_globals([
    # OmegaConf типы
    DictConfig,
    ListConfig,
    AnyNode,
    Metadata,

    # Базовые Python типы
    dict,
    list,
    set,
    tuple,
    int,
    float,
    bool,
    str,
    type(None),

    # Коллекции
    collections.defaultdict,
    collections.OrderedDict,
])

```

```python
tuner = TabularModelTuner(
    data_config=data_config,
    model_config=model_config,
    optimizer_config=optimizer_config,
    trainer_config=trainer_config
)
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    result = tuner.tune(
        train=X_train.join(y_train),
        validation=X_test.join(y_test),
        search_space=search_space,
        strategy="grid_search",
        # cv=5, # Uncomment this to do a 5 fold cross validation
        metric="mean_squared_error",
        mode="min",
        progress_bar=True,
        verbose=False # Make True if you want to log metrics and params each iteration
    )
```

```python
print("Best Score: ", result.best_score)
print(result.best_params)
```

**Подбор размера батчей**

```python
batch_sizes = [32, 64, 128]
batch_results = {}

for batch in batch_sizes:
    trainer_config = TrainerConfig(
        batch_size=batch,
        max_epochs=50,
        early_stopping="valid_loss",
        early_stopping_mode="min",
        early_stopping_patience=5,
        checkpoints="valid_loss",
        load_best=True,
        progress_bar="none",
        trainer_kwargs=dict(enable_model_summary=False)
    )

    model = TabularModel(
        data_config=data_config,
        model_config=model_config,
        optimizer_config=optimizer_config,
        trainer_config=trainer_config
    )

    # Обучение
    model.fit(train=X_train.join(y_train), validation=X_test.join(y_test))

    # Оценка на валидации
    val_metrics = model.evaluate(test=X_test.join(y_test))
    batch_results[batch] = val_metrics[0]["test_mean_squared_error"] ** 0.5  # sqrt(MSE) = RMSE


print("RMSE по batch_size:", batch_results)
```

# Обучение модели на новых гиперпараметрах

```python
class ANN_Model(nn.Module):
  def __init__(self, input_features=4, hidden1=20, hidden2=20, out_features=1,
               activation=nn.LeakyReLU, dropout=0.0, use_batch_norm=True):
    super().__init__()
    self.use_batch_norm = use_batch_norm
    self.dropout = nn.Dropout(dropout) if dropout > 0 else nn.Identity()
    self.activation = activation()

    self.bn1 = nn.BatchNorm1d(hidden1)
    self.bn2 = nn.BatchNorm1d(hidden2)

    self.f_connected1=nn.Linear(input_features, hidden1)
    self.f_connected2=nn.Linear(hidden1, hidden2)
    self.out=nn.Linear(hidden2, out_features)

  def forward(self, x):
    x = self.f_connected1(x)
    x = self.bn1(x)
    x = self.activation(x)
    x = self.dropout(x)

    x = self.f_connected2(x)
    x = self.bn2(x)
    x = self.activation(x)
    x = self.dropout(x)

    x=self.out(x)
    return x

batch_size=64

model = ANN_Model(input_features=X_train_tensor.shape[1],
                  activation=nn.LeakyReLU,
                  dropout=0.0,
                  use_batch_norm=True)

loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)

epochs = 1000
final_losses = []

for i in range(epochs):
  y_pred = model.forward(X_train_tensor)
  loss = loss_function(y_pred, y_train_tensor)
  final_losses.append(loss.item())

  if i%10==1:
    print("Epoch number: {} and the loss: {}".format(i, loss.item()))

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F

#График изменения потерь
plt.figure(figsize=(10,5))
plt.plot(final_losses)
plt.xlabel("Эпохи")
plt.ylabel("MSE Loss")
plt.title("Изменение ошибки на тренировке")
plt.show()

# 2️MSE на тестовой выборке
model.eval()  # переводим модель в режим оценки
with torch.no_grad():  # отключаем градиенты
    y_test_pred = model(X_test_tensor)
    mse_test = F.mse_loss(y_test_pred, y_test_tensor)
    rmse_test = torch.sqrt(mse_test)
    print(f"RMSE на тестовой выборке: {rmse_test.item():.4f}")
```
