# Псевдо-разметка

````python
import numpy as np
from skimage.measure import label, regionprops

def generate_bbox(img, min_area_ratio=0.001):
    if len(img.shape) == 3:
        img = img.mean(axis=2)

    H, W = img.shape

    mean_val = img.mean()
    mask_bright = img > mean_val + 0.3*(img.max()-img.min())
    mask_dark   = img < mean_val - 0.3*(img.max()-img.min())
    mask = mask_bright | mask_dark

    labeled = label(mask)
    regions = regionprops(labeled)
    if len(regions) == 0:
        return [W*0.25, H*0.25, W*0.5, H*0.5]

    max_area = 0
    best_region = None
    for r in regions:
        y1, x1, y2, x2 = r.bbox
        area = (x2 - x1) * (y2 - y1)
        if area >= min_area_ratio * (H*W) and area > max_area:
            max_area = area
            best_region = r

    if best_region is None:
        return [W*0.25, H*0.25, W*0.5, H*0.5]

    y1, x1, y2, x2 = best_region.bbox
    w = x2 - x1
    h = y2 - y1
    return [float(x1), float(y1), float(w), float(h)]

def convert_to_yolo(box, img_w, img_h):
    x_min, y_min, w, h = box
    x_center = x_min + w / 2
    y_center = y_min + h / 2

    return [
        x_center / img_w,
        y_center / img_h,
        w / img_w,
        h / img_h
    ]

def add_bboxes(example):
    img = example["image"]

    w, h = img.size

    img_np = np.array(img)

    raw_bbox = generate_bbox(img_np)
    yolo_bbox = convert_to_yolo(raw_bbox, w, h)

    example["bbox"] = yolo_bbox
    example["label"] = 0
    return example

ds = ds.map(add_bboxes)
print(ds['train'][0])
````

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def show_examples(dataset, num_examples=4):
    indices = [int(i) for i in np.random.choice(len(dataset), size=num_examples, replace=False)]

    fig, axes = plt.subplots(1, num_examples, figsize=(5*num_examples, 5))

    if num_examples == 1:
        axes = [axes]

    for ax, idx in zip(axes, indices):
        example = dataset[idx]
        img = example["image"].convert("RGB")
        w, h = img.size
        xc, yc, bw, bh = example["bbox"]

        box_w = bw * w
        box_h = bh * h
        box_x = xc * w - box_w / 2
        box_y = yc * h - box_h / 2

        ax.imshow(img, cmap='gray')
        rect = patches.Rectangle(
            (box_x, box_y),
            box_w,
            box_h,
            linewidth=2,
            edgecolor='red',
            facecolor='none'
        )
        ax.add_patch(rect)
        ax.axis('off')

    plt.show()

show_examples(ds["train"], num_examples=4)
````

В данном разделе реализована простая процедура псевдо-разметки для датасета Brain-tumor.

1. **Функция `generate_bbox`**:
   - Преобразует изображение в градации серого.
   - Находит яркие и тёмные области относительно среднего значения.
   - Строит маску контрастных областей.
   - Использует связные компоненты (`label`, `regionprops`) для поиска самой крупной области.
   - Возвращает ограничивающую рамку (bbox) этой области.
   - Если область не найдена — создаёт fallback-рамку по центру изображения.

2. **Функция `convert_to_yolo`**:
   - Конвертирует bbox из формата `(x_min, y_min, w, h)` в YOLO-нормированный формат
     `(x_center, y_center, w_norm, h_norm)`.

3. **Функция `add_bboxes`**:
   - Применяет псевдо-разметку к каждому примеру датасета.
   - Генерирует bbox.
   - Конвертирует его в формат YOLO.
   - Добавляет bbox и фиктивный класс `label = 0` (один класс — опухоль).

4. **Обработка датасета**:
   - `ds = ds.map(add_bboxes)` автоматически добавляет bbox ко всем изображениям в датасете.




# Подготовка данных для PyTorch

````python
import torch
from torch.utils.data import DataLoader

def process_example(item, processor):
    img = item["image"].convert("RGB")
    bbox = torch.tensor(item["bbox"], dtype=torch.float32)
    label = torch.tensor(item["label"], dtype=torch.float32)
    target = torch.cat([label.unsqueeze(0), bbox])
    pixel_values = processor(images=img, return_tensors="pt")["pixel_values"].squeeze(0)
    return pixel_values, target

def collate_fn(batch):
    imgs = torch.stack([b[0] for b in batch])
    targets = torch.stack([b[1] for b in batch])
    return imgs, targets

train_list = [process_example(item, processor) for item in ds["train"]]
val_list   = [process_example(item, processor) for item in ds["validation"]]

train_loader = DataLoader(train_list, batch_size=8, shuffle=True, collate_fn=collate_fn)
val_loader   = DataLoader(val_list, batch_size=8, collate_fn=collate_fn)

imgs, targets = next(iter(train_loader))
print(imgs.shape)
print(targets.shape)
````

1. **Функция `process_example`**:
   - Конвертирует изображение в RGB и пропускает через предобработчик (`processor`), связанный с выбранным backbone.
   - Превращает bbox и label в тензоры PyTorch.
   - Формирует целевой вектор вида `[class, x_center, y_center, w, h]`.
   - Возвращает пару: (обработанное изображение, целевой вектор).

2. **Функция `collate_fn`**:
   - Объединяет изображения и цели в батч.
   - Возвращает два тензора: `imgs` и `targets`.

3. **Создание списков данных `train_list`, `val_list`**:

4. **DataLoader**:

5. **Проверка**:
   - Печать форм батча подтверждает:
     - `imgs`: `[batch, channels, height, width] = [8, 3, 224, 224]`

6. **Блок визуализации bbox**:
   - Берёт 4 случайных изображения из батча.

   - Рисует YOLO-рамку красным прямоугольником.
   - Проверяет корректность псевдо-разметки.



# Реализация YOLO-head

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class YOLOHead(nn.Module):
    def __init__(self, in_channels, grid_size=7, num_classes=1):
        super().__init__()
        self.S = grid_size
        self.num_classes = num_classes

        self.conv = nn.Conv2d(in_channels, self.S*self.S*(5+self.num_classes), kernel_size=1)

    def forward(self, features):
        x = self.conv(features)
        x = F.adaptive_avg_pool2d(x, (self.S, self.S))
        x = x.flatten(2)
        x = x.mean(-1)
        x = x.view(-1, self.S, self.S, 5+self.num_classes)
        return x
````

В этом разделе реализуется упрощённая голова YOLO, которая принимает фичи от предобученного backbone и предсказывает параметры объектов в каждой ячейке сетки.

1. **Архитектура головы**
   - Вход: фичи из backbone размера `[B, C, H', W']`.
   - Свертка `1×1` уменьшает число каналов до:
     ```
     S * S * (5 + num_classes)
     ```
     где:
     - `S × S` — сетка YOLO,
     - `5` — параметры бокса `(x, y, w, h, objectness)`,
     - `num_classes` — количество классов (здесь = 1).

2. **Приведение размеров**
   - `adaptive_avg_pool2d` сжимает spatial размер фичей к `(S, S)` для обеспечения одинаковой сетки.
   - Затем тензор приводится к форме `[B, S, S, 5 + num_classes]`.

3. **Интерпретация вывода**
   - Выход головы — predictions YOLO для каждой клетки сетки.
   - На каждые `S × S` ячеек модель предсказывает bounding box и вероятность.



# Реализация Loss функции

````python
def yolo_loss(preds, targets):
    B, S, _, _ = preds.shape

    pred_bbox = preds[:, S//2, S//2, :4]
    pred_cls  = preds[:, S//2, S//2, 5:]

    target_bbox = targets[:, 1:]
    target_cls  = targets[:, 0].long()

    loss_bbox = F.mse_loss(pred_bbox, target_bbox)

    loss_cls = F.cross_entropy(pred_cls, target_cls)

    return loss_bbox + loss_cls
````

# Обучаем

````python
optimizer = torch.optim.Adam(list(backbone.parameters()), lr=1e-4)
yolo_head = YOLOHead(in_channels=backbone.config.hidden_sizes[-1]).cuda()
backbone.cuda()

for epoch in range(6):
    for imgs, targets in train_loader:
        imgs = imgs.cuda()
        targets = targets.cuda()

        features = backbone(imgs).last_hidden_state

        preds = yolo_head(features)
        loss = yolo_loss(preds, targets)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch} Loss: {loss.item():.6f}")
````

# Реализация NMS

````python
def bbox_iou(box1, box2):
    x1 = torch.max(box1[:,0], box2[:,0])
    y1 = torch.max(box1[:,1], box2[:,1])
    x2 = torch.min(box1[:,2], box2[:,2])
    y2 = torch.min(box1[:,3], box2[:,3])

    inter = (x2 - x1).clamp(0) * (y2 - y1).clamp(0)
    area1 = (box1[:,2]-box1[:,0]) * (box1[:,3]-box1[:,1])
    area2 = (box2[:,2]-box2[:,0]) * (box2[:,3]-box2[:,1])
    union = area1 + area2 - inter
    return inter / (union + 1e-6)

def nms(boxes, scores, iou_threshold=0.5):
    if boxes.numel() == 0:
        return torch.tensor([], dtype=torch.long)

    idxs = scores.argsort(descending=True)
    keep = []
    while idxs.numel() > 0:
        i = idxs[0].item()
        keep.append(i)
        if idxs.numel() == 1:
            break

        ious = bbox_iou(boxes[i:i+1], boxes[idxs[1:]]).squeeze(0)
        idxs = idxs[1:][ious < iou_threshold]
    return torch.tensor(keep, dtype=torch.long)
````

В этом блоке реализованы две ключевые функции

- вычисление IoU (Intersection over Union),
- алгоритм NMS (Non-Maximum Suppression).



**Функция `bbox_iou`**

* Что делает:*
- Получает **box1** и **box2** в формате `[x1, y1, x2, y2]`.
- Находит пересечение прямоугольников.
- Вычисляет площади.
- Возвращает IoU = inter / union.



**Функция `nms`**

*Что делает:*
- Принимает bbox-ы и их confidence-scores.
- Сортирует боксы по score (по убыванию).
- Итеративно:
  - оставляет бокс с максимальным score,
  - удаляет все боксы, у которых IoU с ним выше порога (`iou_threshold`).



````python
import matplotlib.pyplot as plt
from torchvision.transforms.functional import to_pil_image

imgs, targets = next(iter(val_loader))
imgs_cuda = imgs.cuda()

features = backbone(imgs_cuda).last_hidden_state
preds = yolo_head(features)

pred_bbox = preds[:, preds.shape[1]//2, preds.shape[2]//2, :4]
pred_obj  = torch.ones(pred_bbox.size(0))

bboxes_abs = []
for i in range(pred_bbox.size(0)):
    x_c, y_c, w, h = pred_bbox[i]
    x1 = (x_c - w/2) * imgs.shape[3]
    y1 = (y_c - h/2) * imgs.shape[2]
    x2 = (x_c + w/2) * imgs.shape[3]
    y2 = (y_c + h/2) * imgs.shape[2]
    bboxes_abs.append([x1, y1, x2, y2])
bboxes_abs = torch.tensor(bboxes_abs)


keep = nms(bboxes_abs, pred_obj, iou_threshold=0.5)
bboxes_final = bboxes_abs[keep]
````

**В этом блоке производится генерация предсказаний модели и применение NMS к выходу YOLO-head**

````python
import torch

conf_threshold = 0.8  # порог уверенности

imgs, targets = next(iter(val_loader))
imgs_cuda = imgs.cuda()


features = backbone(imgs_cuda).last_hidden_state
preds = yolo_head(features)

B, S, _, _ = preds.shape

bboxes_abs_list = []

for b in range(B):
    bboxes = []
    scores = []

    H, W = imgs.shape[2], imgs.shape[3]

    for i in range(S):
        for j in range(S):
            x_c, y_c, w_box, h_box = torch.sigmoid(preds[b, i, j, :4])
            obj_conf = torch.sigmoid(preds[b, i, j, 4])

            if obj_conf < conf_threshold:
                continue

            x1 = (x_c - w_box/2) * W
            y1 = (y_c - h_box/2) * H
            x2 = (x_c + w_box/2) * W
            y2 = (y_c + h_box/2) * H

            bboxes.append([x1, y1, x2, y2])
            scores.append(obj_conf)

    if len(bboxes) == 0:
        bboxes_abs_list.append(torch.tensor([[W*0.25, H*0.25, W*0.75, H*0.75]]))
        continue

    bboxes = torch.tensor(bboxes)
    scores = torch.tensor(scores)

    keep = nms(bboxes, scores, iou_threshold=0.5)
    bboxes_final = bboxes[keep]

    bboxes_abs_list.append(bboxes_final)
````

Выводы к этому блоку кода

- **Фильтрация по confidence**: добавлен порог `conf_threshold = 0.8`, чтобы отбрасывать слабые предсказания. Это уменьшает количество ложных положительных bbox.  
- **Преобразование координат**: предсказанные нормализованные bbox `(x_center, y_center, w, h)` переводятся в абсолютные координаты изображения.  
- **Fallback**: если ни один bbox не прошёл порог уверенности, создаётся центральный bbox на изображении (`[0.25*W, 0.25*H, 0.75*W, 0.75*H]`), чтобы гарантировать наличие bbox.  
- **Применение NMS**: оставшиеся bbox после порога фильтруются функцией `nms` для устранения перекрывающихся предсказаний.  
- **Результат**: `bboxes_abs_list` содержит финальные bbox для каждого изображения в батче, готовые для визуализации или вычисления метрик.

# Метрики

````python
import torch

def bbox_iou(box1, box2):
    N = box1.size(0)
    M = box2.size(0)

    box1 = box1.unsqueeze(1).expand(N, M, 4)
    box2 = box2.unsqueeze(0).expand(N, M, 4)

    x1 = torch.max(box1[:,:,0], box2[:,:,0])
    y1 = torch.max(box1[:,:,1], box2[:,:,1])
    x2 = torch.min(box1[:,:,2], box2[:,:,2])
    y2 = torch.min(box1[:,:,3], box2[:,:,3])

    inter = (x2 - x1).clamp(0) * (y2 - y1).clamp(0)
    area1 = (box1[:,:,2]-box1[:,:,0]) * (box1[:,:,3]-box1[:,:,1])
    area2 = (box2[:,:,2]-box2[:,:,0]) * (box2[:,:,3]-box2[:,:,1])
    union = area1 + area2 - inter
    return inter / (union + 1e-6)

def compute_metrics_no_threshold(preds, targets, iou_threshold=0.5, img_w=224, img_h=224):
    B, S, _, _ = preds.shape
    precisions, recalls, ious_list = [], [], []

    for b in range(B):
        grid_bboxes = preds[b,:,:,:4]
        grid_obj   = torch.ones(S*S)

        bboxes_abs = []
        for i in range(S):
            for j in range(S):
                x_c, y_c, w_box, h_box = torch.sigmoid(grid_bboxes[i,j])
                x1 = (x_c - w_box/2) * img_w
                y1 = (y_c - h_box/2) * img_h
                x2 = (x_c + w_box/2) * img_w
                y2 = (y_c + h_box/2) * img_h
                bboxes_abs.append([x1, y1, x2, y2])
        bboxes_abs = torch.tensor(bboxes_abs)

        keep = nms(bboxes_abs, grid_obj, iou_threshold=iou_threshold)
        pred_boxes_final = bboxes_abs[keep]

        tx, ty, tw, th = targets[b,1:]
        x1_true = (tx - tw/2) * img_w
        y1_true = (ty - th/2) * img_h
        x2_true = (tx + tw/2) * img_w
        y2_true = (ty + th/2) * img_h
        true_box = torch.tensor([[x1_true, y1_true, x2_true, y2_true]])

        if pred_boxes_final.numel() == 0:
            ious_list.append(0)
            precisions.append(0)
            recalls.append(0)
            continue

        ious = bbox_iou(pred_boxes_final, true_box)
        max_iou = ious.max().item()
        ious_list.append(max_iou)

        precisions.append(1 if max_iou >= iou_threshold else 0)
        recalls.append(1 if max_iou >= iou_threshold else 0)

    mean_iou = sum(ious_list)/len(ious_list)
    precision = sum(precisions)/len(precisions)
    recall = sum(recalls)/len(recalls)

    return mean_iou, precision, recall

imgs, targets = next(iter(val_loader))
imgs_cuda = imgs.cuda()
features = backbone(imgs_cuda).last_hidden_state
preds = yolo_head(features)

mean_iou, precision, recall = compute_metrics_no_threshold(
    preds.cpu(), targets.cpu(),
    img_w=imgs.shape[3], img_h=imgs.shape[2]
)
print(f"IoU: {mean_iou:.3f}, Precision: {precision:.3f}, Recall: {recall:.3f}")
````



- **Функция `bbox_iou`**: вычисляет матрицу IoU между всеми парами предсказанных и истинных боксов. IoU показывает степень перекрытия между предсказанным и реальным объектом.  

- **Функция `compute_metrics`**: оценивает качество предсказаний YOLO на батче изображений:
  - Переводит bbox из нормализованных координат YOLO в абсолютные координаты пикселей.
  - Применяет NMS для удаления перекрывающихся предсказаний.
  - Сравнивает каждый предсказанный bbox с истинным и вычисляет максимальное IoU.
  - На основе IoU и заданного порога (`iou_threshold=0.5`) считает precision и recall как бинарные показатели (1 если объект найден, 0 если нет).

- **Применение к батчу**:
  - Берутся изображения и таргеты из `val_loader`.
  - Генерируются предсказания через backbone + YOLO head.
  - Вычисляются метрики: средний IoU, точность и полнота.
